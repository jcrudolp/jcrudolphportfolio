## Build and Evaluate at least one Model:


```python
# Set a variable X to the features of the dataframe and y to the target column.
X = df.drop('output', axis=1)
y = df['output']
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the training data and transform it
X_train = scaler.fit_transform(X_train)

# Transform the test data using the scaler fitted on the training data
X_test = scaler.transform(X_test)
```

```python
## Simple Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Instantiating the object
logmodel = LogisticRegression(max_iter=300)

# Fit the model
logmodel.fit(X_train, y_train)

# Predict values for the testing data
log_pred = logmodel.predict(X_test)

# Create a confusion matrix and classification report for the model
print(confusion_matrix(y_test, log_pred))
print(classification_report(y_test, log_pred))
```

```python
## Decision Tree Classifier 

from sklearn.tree import DecisionTreeClassifier

# Instantiating the object
dt = DecisionTreeClassifier(random_state = 42)

# Fit the model
dt.fit(X_train, y_train)

# Predict values for the testing data
dt_pred = dt.predict(X_test)

# Create a confusion matrix and classification report for the model
print(confusion_matrix(y_test, dt_pred))
print(classification_report(y_test, dt_pred))
```

```python
## Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

# Instantiating the object
rf = RandomForestClassifier()

# Fit the model
rf.fit(X_train, y_train)

# Predict values for the testing data
rf_pred = dt.predict(X_test)

# Create a confusion matrix and classification report for the model
print(confusion_matrix(y_test, rf_pred))
print(classification_report(y_test, rf_pred))
```

```python
## L2 regularized logistic regression with cross-validation

from sklearn.linear_model import LogisticRegressionCV

# L2 regularized logistic regression with cross-validation
lr_l2 = LogisticRegressionCV(Cs=20, cv=4, penalty='l2').fit(X_train, y_train)

# Predict values for the testing data
l2_pred = lr_l2.predict(X_test)

# Create a confusion matrix and classification report for the model
print(confusion_matrix(y_test, l2_pred))
print(classification_report(y_test, l2_pred))
```

```python
# Support Vector Machine 
## Simple SVC Model

from sklearn.svm import SVC

# Create and fit a SVC model
svc = SVC()
svc.fit(X_train, y_train.values.ravel())

# Make predictions on the test data
svc_pred = svc.predict(X_test)

# Create a confusion matrix and classification report for the model
print(confusion_matrix(y_test, svc_pred))
print(classification_report(y_test, svc_pred))
```

```python
## Searching for Best SVC Parameters:

from sklearn.model_selection import GridSearchCV

# Define the grid of parameters to search
params_grid = {'C': [0.05, 0.1, 0.5, 1, 10],
               'kernel': ['poly', 'rbf', 'sigmoid']
                }

# Create another SVC model
model = SVC()

# Define a GridSearchCV to search for the best parameters
grid_search = GridSearchCV(estimator = model, 
                           param_grid = params_grid, 
                           scoring='f1',
                           cv = 5, verbose = 1)

# Search the best parameters with training data
grid_search.fit(X_train, y_train.values.ravel())

# Get the best parameters found by GridSearchCV
best_params = grid_search.best_params_
best_params
```

```python
## SVC using Best Parameters:

# Retraining an SVC model with the best parameters
best_svc = SVC(C=0.1, kernel='sigmoid')
best_svc.fit(X_train, y_train.values.ravel())

# Make predictions on the test data
svc_pred = best_svc.predict(X_test)

# Create a confusion matrix and classification report for the model
print(confusion_matrix(y_test, svc_pred))
print(classification_report(y_test, svc_pred))
```
